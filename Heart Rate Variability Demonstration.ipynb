# Heart Rate Variability Calculations

## Time Domain Calculations

```python
# Basic packages to import
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
import statistics
```

```python
# Read a Text File containing R-R Intervals of a Mouse. Data is exported from ecgAUTO software.
rr_frame = pd.read_csv("c:\data\hr_export_1.txt")
print(rr_frame)
```

```python
# Extract the column for convenience
rr_int = rr_frame["RR(ms)"]
```

```python
# Plot the RR Interval Data Set
plt.plot(rr_int)
plt.xlabel("Beat Number")
plt.ylabel("RR Interval, ms")
plt.axis([0, 80000, 0, 300])
plt.grid(True)
plt.show()
```

```python
# Compute the Mean of the RR Intervals
rr_mean = statistics.mean(rr_int)
print("Mean R-R' Interval: ", round(rr_mean,1), "ms")
```

```python
# Compute the Standard Deviation of the RR Intervals
sdrr = statistics.stdev(rr_int)
print("Std Deviation of RR Intervals (SDRR):", round(sdrr,2), "ms")
```

```python
# Compute the RMSSD
# Root mean square of successive RR interval differences
i = 1
max = len(rr_int)
dev_sq = []
while i < max:
    diff_sq = (rr_int[i] - rr_int[i-1])**2
    dev_sq.append(diff_sq)
    i += 1
rmssd = np.sqrt(np.mean(dev_sq))
print ("RMSSD:", round(rmssd,1), "ms")
```

```python
# Plot the Histogram of RR intervals
plt.hist(rr_int, 100, density=True, facecolor='b', alpha=0.75)
plt.title("RR Interval Hitogram")
plt.xlabel("R-R' Interval, ms")
plt.ylabel("Density")
plt.axis([60, 120, 0, 0.2])
plt.grid(True)
plt.show()
```
